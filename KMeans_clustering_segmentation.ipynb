```python
import cv2
import imageio
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=imageio.imread("mountain.jpg")
```

```python
plt.imshow(img.astype(np.uint8))
```

```python
img=img.astype(np.float32)
```

```python
img.dtype
```

```python
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 1000, 0.2)
```

```python
criteria
```

```python
K=3
```

```python
ret,label,center=cv2.kmeans(img,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
```

```python
label.flatten()
```

```python
center=center.astype(np.uint8)
```

```python
center
```

```python
ret
```

```python
segmented_img=center[label.flatten()]
```

```python
segmented_img=segmented_img.reshape(img.shape)
```

```python
plt.imshow(segmented_img) # 
```

```python
label=label.flatten()
```

```python
segmented_img[label!=0]=[0,0,0]
```

```python
plt.imshow(segmented_img) # Masked Image
```
