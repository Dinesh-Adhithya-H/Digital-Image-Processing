```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
import time
```

```python
def DFT_1D(f):
  F=np.zeros(f.shape,dtype=np.complex64)
  n=f.shape[0]
  for u in np.arange(n):
    for x in np.arange(n):
      F[u]+=f[x]*np.exp(-1j*2*np.pi*u*x/n)
  return F/np.sqrt(n)
```

```python
t=np.arange(0,1,0.001)
f=np.sin(2*np.pi*t)+0.5*np.sin(2*np.pi*t*4)+0.6*np.cos(2*np.pi*t*18)+0.3*np.cos(4*2*np.pi*t)
plt.plot(t,f)
```

```python
F=DFT_1D(f)
fq=np.arange(1000)
plt.plot(fq,F)
plt.show()
```

```python
def DFT_2D(f):
  F=np.zeros(f.shape,dtype=np.complex64)
  n,m=f.shape[0:2]

  start=time.time()

  x=np.arange(n)
  y=np.arange(m)
  for u in np.arange(n):
    for v in np.arange(m):
      for x in np.arange(n):
        for y in np.arange(m):
          F[u,v]+=f[x,y]*np.exp(-1j*2*np.pi*(u*x/n+v*y/m))
  end=time.time()
  elapsed=end-start
  print(elapsed)
  return F/np.sqrt(n*m)
```

```python
img1_sub=img1[100:132,100:132,0]
f1s=DFT_2D(img1_sub)
```
