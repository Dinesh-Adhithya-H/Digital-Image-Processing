 # Frequecy Analysis

```python
t=np.arange(0,1,0.001)
```

```python
t.shape
```

```python
sine=np.sin(2*np.pi*t*4)
plt.figure(figsize=(5,5))
plt.plot(t,sine)
```

```python
cose=np.cos(2*np.pi*t*18)
plt.figure(figsize=(5,5))
plt.plot(t,cose)
```

```python
sine2=np.sin(2*np.pi*t)
cose2=np.cos(2*np.pi*t*4)
```

```python
fun=sine+sine2+cose+cose2
plt.plot(t,fun)
```

```python
omega=4
match_sin=fun*np.sin(omega*2*np.pi*t)
match_cos=fun*np.cos(omega*2*np.pi*t)
```
