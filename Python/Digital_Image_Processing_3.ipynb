# Smoothing

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img=cv2.imread(r"C:\Users\hp\Downloads\cr7.jpg",0)
```

```python
kernel=np.ones((5,5),np.float64)/25
```

```python
dst=cv2.filter2D(img,-1,kernel)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.title("original")
plt.imshow(img,cmap="gray")
plt.subplot(122)
plt.title("averaging")
plt.imshow(dst,cmap="gray")
plt.show()
```

```python
dst=cv2.blur(img,(5,5))
```

```python
plt.imshow(dst,cmap="gray")
```

# Gaussian Filter

```python
dst=cv2.GaussianBlur(img,(5,5),0)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.title("original")
plt.imshow(img,cmap="gray")
plt.subplot(122)
plt.title("gaussian smoothing")
plt.imshow(dst,cmap="gray")
plt.show()
```

# Median Filter

```python
dst=cv2.medianBlur(img,3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.title("original")
plt.imshow(img,cmap="gray")
plt.subplot(122)
plt.title("median blur")
plt.imshow(dst,cmap="gray")
plt.show()
```
