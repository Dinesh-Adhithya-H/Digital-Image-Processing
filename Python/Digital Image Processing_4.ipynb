```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
gray=cv2.imread(r"C:\Users\hp\Downloads\blk_to_white.png",0)
```

```python
plt.imshow(gray,cmap="gray")
plt.show()
```

```python
plt.plot(gray[0,:])
plt.show()
```

```python
filter_=np.array([[0,1,0],[1,0,-1],[0,-1,0]])
```

```python
new_img = cv2.filter2D(src=gray, ddepth=-1, kernel=filter_)
```

```python
plt.imshow(new_img,cmap="gray")
```

```python
plt.plot(new_img[0,:])
plt.show()
```

```python
laplacian=np.array([[0,1,0],[1,-4,1],[0,1,0]])
```

```python
new_img1 = cv2.filter2D(src=gray, ddepth=-1, kernel=laplacian)
```

```python
plt.imshow(new_img1,cmap="gray")
```

# Edge sharpening using laplacian

```python
mask=lungs-lungs_lap
plt.imshow(lungs+0.21*mask,cmap="gray")
plt.show()
```

# Unsharp masking 

```python
plt.imshow(lungs-0.1*cv2.GaussianBlur(lungs,(5,5),2.5),cmap="gray")
plt.show()
```
