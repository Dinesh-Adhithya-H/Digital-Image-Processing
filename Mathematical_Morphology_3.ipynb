```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.util import img_as_ubyte
```

```python
plt.imshow(data.shepp_logan_phantom(),cmap="gray")
```

```python
img=data.shepp_logan_phantom()
```

```python
def plot_images(ori,filt):
  plt.subplot(1,2,1)
  plt.title("original image")
  plt.imshow(ori,cmap="gray")

  plt.subplot(1,2,2)
  plt.title("processed image")
  plt.imshow(filt,cmap="gray")
```

```python
plot_images(img,img)
```

```python
from skimage.morphology import (erosion,dilation,opening,closing,white_tophat)
from skimage.morphology import black_tophat, skeletonize, convex_hull_image ,disk
```

```python
foot=disk(5)
eroded=erosion(img,foot)
plot_images(img,eroded)
```

```python
dilated = dilation(img,foot)
plot_images(img,dilated)
```

```python
opened=opening(img,foot)
plot_images(img,opened)
```

```python
img[10:30,200:210]=0.0
closed=closing(img,foot)
plot_images(img,closed)
```

```python
w_tophat=white_tophat(img,foot)
plot_images(img,w_tophat)
```

```python
b_tophat=black_tophat(img,foot)
plot_images(img,b_tophat)
```

```python
horse=data.horse()
sk=skeletonize(horse==0)

plot_images(horse,sk)
```

```python
c_hull=convex_hull_image(horse==0)

plot_images(horse,c_hull)
```
